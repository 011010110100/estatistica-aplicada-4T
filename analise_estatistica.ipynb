```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Lendo o arquivo csv e passando para o pandas
df = pd.read_csv('./Dados/Anexo_Projeto_fifa_world_national_teams_versão_oficial_20241.csv', sep=';')

# Pegando as primeiras linhas do arquivo para verificar os dados
df.head()
```

```python
def mostrar_dados(df, lista, num_max_linhas=5):
    """
    Percorre o DataFrame e retorna os valores das colunas especificadas na lista.

    Argumentos:
        df: O DataFrame a ser percorrido.
        colunas (lista): Uma lista com os nomes das colunas a serem mostradas.
        num_max_linhas (int, opcional): O número máximo de linhas a serem mostradas.
    """

    for i, row in df.iterrows():
        if i >= num_max_linhas: # Interrompe o loop
            break  
        print(f"Índice: {i}")
        
        for coluna in lista:
            print(f"  {coluna}: {row[coluna]}")
        print("-" * 20)

# 1. Encontre e liste uma variável de cada tipo (qualitativa nominal, qualitativa ordinal, quantitativa discreta e quantitativa contínua); 

# qualitativa nominal: Rotulam Algo.
# Por exemplo: club_team (representa o time do jogador, não há ordem)

# qualitativa ordinal: existe uma ordem clara
# Por exemplo: skill_moves (representam a habilidade do jogador de 1 a 5)

# quantitativa discreta: quando podemos contar.
# Por exemplo: Atacante (se o jogador é atacante ou não)

# quantitativa contínua: posui um intervalo.
# Por exemplo: height_cm (representa a altura em centímetros, podendo assumir valors contínuos dentro de um intervalo)
colunas_desejadas = ['club_team', 'skill_moves(1-5)', 'Atacante', 'height_cm']

mostrar_dados(df, colunas_desejadas, 2)
```
